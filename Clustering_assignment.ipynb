```python
from pycaret.datasets import get_data
from pycaret.clustering import *

def eval(data,method):
  iplDataSet = get_data(data, verbose=False)
  setup(data = iplDataSet, verbose=False)

  print(" {} method without any data preprocessing".format(method))
  print("For Cluster = 3")
  x = create_model(method, num_clusters = 3)

  print("For Cluster = 4")
  x = create_model(method, num_clusters = 4)

  print("For Cluster = 5")
  x = create_model(method, num_clusters = 5)

  print("For Cluster = 6")
  x = create_model(method, num_clusters = 6)
  print("--------------------------------------------------------------")

  print(" {} method with nor ".format(method))
  setup(data = iplDataSet, normalize = True, normalize_method = 'minmax', verbose=False)

  print("For Cluster = 3")
  x = create_model(method, num_clusters = 3)

  print("For Cluster = 4")
  x = create_model(method, num_clusters = 4)

  print("For Cluster = 5")
  x = create_model(method, num_clusters = 5)

  print("For Cluster = 6")
  x = create_model(method, num_clusters = 6)

  print("--------------------------------------------------------------")
  print(" {} method with trans ".format(method))
  setup(data = iplDataSet, transformation = True, transformation_method = 'yeo-johnson', verbose=False)

  print("For Cluster = 3")
  x = create_model(method, num_clusters = 3)

  print("For Cluster = 4")
  x = create_model(method, num_clusters = 4)

  print("For Cluster = 5")
  x = create_model(method, num_clusters = 5)

  print("For Cluster = 6")
  x = create_model(method, num_clusters = 6)

  print("--------------------------------------------------------------")
  print(" {} method with pca".format(method))
  setup(data = iplDataSet, pca = True, pca_method = 'kernel', verbose=False)

  print("For Cluster = 3")
  x = create_model(method, num_clusters = 3)

  print("For Cluster = 4")
  x = create_model(method, num_clusters = 4)

  print("For Cluster = 5")
  x = create_model(method, num_clusters = 5)

  print("For Cluster = 6")
  x = create_model(method, num_clusters = 6)

  print("--------------------------------------------------------------")
  print(" {} method with trans + nor ".format(method))
  setup(data = iplDataSet, transformation = True, normalize = True,
      normalize_method = 'zscore', transformation_method = 'yeo-johnson', verbose=False)

  print("For Cluster = 3")
  x = create_model(method, num_clusters = 3)

  print("For Cluster = 4")
  x = create_model(method, num_clusters = 4)

  print("For Cluster = 5")
  x = create_model(method, num_clusters = 5)

  print("For Cluster = 6")
  x = create_model(method, num_clusters = 6)
  print("--------------------------------------------------------------")

  print(" {} method with trans + nor + pca".format(method))
  setup(data = iplDataSet, transformation = True, normalize = True, pca = True, normalize_method = 'zscore',transformation_method = 'yeo-johnson',pca_method = 'kernel', verbose=False)

  print("For Cluster = 3")
  x = create_model(method, num_clusters = 3)

  print("For Cluster = 4")
  x = create_model(method, num_clusters = 4)

  print("For Cluster = 5")
  x = create_model(method, num_clusters = 5)

  print("For Cluster = 6")
  x = create_model(method, num_clusters = 6)
  print("--------------------------------------------------------------")
```

```python
eval(data = "mice",method = 'kmeans')
```
